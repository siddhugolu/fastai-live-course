Getting a feel of the dataset - it's a multiclassification problem, keep in mind.

```python
print(path)
```

```python
df = pd.read_csv(path/'train_v2.csv')
```

```python
df.head()
```

```python
tfms = get_transforms(flip_vert=True, max_lighting=0.1, max_zoom=1.05, max_warp=0.)

#flip_vert = True here indicates that flipping the images vertically would cause no change in the satellite images.
# As a rule of thumb, doing this makes the model more generalizable.
```

```python
np.random.seed(42)
src = (ImageItemList.from_csv(path, 'train_v2.csv', folder='train-jpg', suffix='.jpg')
      .random_split_by_pct(0.2)
      .label_from_df(sep = ' '))
```

```python
data = (src.transform(tfms, size = 128)
       .databunch().normalize(imagenet_stats))
```

```python
data.show_batch(rows = 3, figsize = (12,9))
```

```python
arch = models.resnet50
```

```python
fbeta??
```

A partial function is used to derive from a function with x arguments, a new function having fewer arguments and the rest of the arguments getting a fixed value.

```python
acc_02 = partial(accuracy_thresh, thresh=0.2)
f_score = partial(fbeta, thresh = 0.2)
learn = create_cnn(data, arch, metrics = [acc_02, f_score])
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr = 0.01
```

```python
learn.fit_one_cycle(5, slice(lr))
```

```python
learn.save('stage-1-rn50')
```

Now fine-tune the whole model

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, slice(1e-5, lr/5))
```

```python
learn.save('stage-2-rn50')
```

Now looking at our databunch that we created earlier, we see that the transformation size was kept as size = 128, 
while in the original Kaggle competition, the size was 256. Hence, we can use transfer learning to further fine-tune our model while using a new databunch whose data will be of size = 256 and see whether we're able to increase the accuracy or not.

```python
data = (src.transform(tfms, size = 256)
       .databunch().normalize(imagenet_stats))

learn.data = data # We're using the same learner but with the new databunch
data.train_ds[0][0].shape
```

We can see that indeed, our image have been resized to 256

```python
learn.freeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```
