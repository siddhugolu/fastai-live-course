The following function searches for the equivalent masked files in the labels directory

```python
get_y_fn = lambda x: path_label/f'{x.stem}_P{x.suffix}'
```

```python
mask = open_mask(get_y_fn(fnames[0]))
```

```python
mask.show(figsize = (8,8), alpha = 1)
```

```python
src_size = np.array(mask.shape[1:])
src_size, mask.data
```

```python
codes = np.loadtxt(path/'codes.txt', dtype = str)
print(codes)
```

# Creating a databunch

```python
size = src_size//2
bs = 8
print(size)
```

```python
src = (SegmentationItemList.from_folder(path_img)
      .split_by_fname_file('../valid.txt')
      .label_from_func(get_y_fn, classes = codes))
```

```python
data = (src.transform(get_transforms(), size = size, tfm_y = True)
       .databunch(bs = bs)
       .normalize(imagenet_stats))
```

```python
data.show_batch(2, figsize = (9,9))
```

```python
data.show_batch(2, figsize = (9,9), ds_type = DatasetType.Valid)
```

# Model

```python
name2id = {v:k for k,v in enumerate(codes)}
void_code = name2id['Void']
 
def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
learn = Learner.create_unet(data, models.resnet18, metrics = [acc_camvid])
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr = 1e-02
```

```python
learn.fit_one_cycle(10, slice(lr))
```

```python
learn.save('stage-1')
```

```python
learn.load('stage-1')
```

```python
learn.show_results()
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
lr_slice = slice(1e-4, lr/5)
```

```python
learn.fit_one_cycle(12, lr_slice)
```

```python
learn.recorder.plot_losses()
```

```python
learn.recorder.plot_lr()
```

```python
learn.save('stage-2')
```
